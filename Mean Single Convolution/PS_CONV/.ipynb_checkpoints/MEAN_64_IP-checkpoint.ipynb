```python
import numpy as np

import cv2
import matplotlib.pyplot as plt
from pynq import Overlay

overlay = Overlay('mean_64.bit')
```

```python
Lap = overlay.mean_0
```

```python
img = cv2.imread('lena64.png', 0)
```

```python
img.shape
```

```python
plt.figure(figsize=(4, 4))
plt.imshow(img, 'gray')
```

```python
%%time
for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        Lap.write(0x4000 + 4 * (i * img.shape[1] + j), int(img[i][j]))

result = img.copy()

for i in range(img.shape[0]):
    for j in range(img.shape[1]):
        result[i][j] = Lap.read(0x8000 + 4 * (i * img.shape[1] + j))
```

```python
plt.figure(figsize=(4, 4))
plt.imshow(result, 'gray')
```
